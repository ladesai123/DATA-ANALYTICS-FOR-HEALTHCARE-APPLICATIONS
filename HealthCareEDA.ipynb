```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



columns = ['Pregnancies', 'Glucose', 'BloddPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv('pima-indians-diabetes.csv', names= columns)
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,10))
for i, column in enumerate(df.columns[:-1]):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Analyzing patient demographics (Age and Pregnancies)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Distribution of Age')

plt.subplot(1, 2, 2)


sns.histplot(df['Pregnancies'], kde=True, bins=10)
plt.title('Distribution of Pregnancies')

plt.tight_layout()
plt.show()
```

```python
#Scatterplot matrix to identify trends

sns.pairplot(df, diag_kind='kde', markers='+', plot_kws={'alpha': 0.3})
plt.suptitle('Scatter Plot Matrix',y = 1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose Levels by Outcome')
plt.xlabel('Diabetes Outcome')
plt.ylabel('Glucose Level')
plt.show()
```

```python
#Analyzing the relationship between BMI and Outcome

plt.figure(figsize=(8,6))
sns.boxplot(x='Outcome', y='BMI', data=df)
plt.title('BMI Levels by Outcome')
plt.xlabel('Diabetes Outcome ')
plt.ylabel('BMI Level')
plt.show()
```
